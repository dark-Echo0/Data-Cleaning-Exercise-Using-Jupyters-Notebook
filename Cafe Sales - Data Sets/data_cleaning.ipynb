## Import Package

```python
import pandas as pd
import numpy as np
```

## Show the first 10 row of the data set

```python
df = pd.read_csv('sales.csv')
df.head(10)
```

## Show the Data type and Non- null Count

```python
df.info()
```

## Count the null values

```python
df.isna().sum()
```

```python
df.describe(include="all")
```

```python
df.isna().sum().sum()
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()

df
```

## Removing Values and Changing Data Types

```python
pd.Series(df["Transaction ID"].unique().tolist()).str.startswith("TXN").value_counts()
```

```python
df
```

```python
df["Transaction Date"].unique()
```

## Check the Unkown and Error Value

```python
df.isin(["UNKNOWN", "ERROR"]).sum()
```

```python
(df == "ERROR").sum()
```

## Remove the Unkown and Error Value

```python
df = df[~df.isin(["UNKNOWN", "ERROR"]).any(axis=1)]
df
```

## Modify the Data types

```python
#show the data types of the column
df.dtypes
```

```python
# Modify the data types corresponding the data in the column example "Quantity" from str to intiger
df["Quantity"] = df["Quantity"].astype(int)
```

```python
df.dtypes
```

```python
#Import Warnings to avoid warnings from the code
import warnings
warnings.filterwarnings("ignore")
```

```python
df
```

```python
df["Price Per Unit"] = df["Price Per Unit"].astype(float)
df["Total Spent"] = df["Total Spent"].astype(float)
```

```python
df.dtypes
```

```python
df["Transaction Date"] = pd.to_datetime(df["Transaction Date"])
```

```python
df.dtypes
```

```python
df.sort_values("Transaction ID").reset_index(drop=True).index
```

```python
df = df.sort_values("Transaction ID").reset_index(drop=True)
df
```

## Save the cleaned data frame to .csv file

```python
df.to_csv("sales_cleaned.csv", index=False)
```

```python
df.to_excel("sales_cleaned_excel.xlsx", index=False)
```
